```python
from bs4 import BeautifulSoup
import requests
```

```python
link = 'https://id.wikipedia.org/wiki/Daftar_film_terlaris'
page = requests.get(link)
soup = BeautifulSoup(page.text, 'html')
```

```python
tabel = soup.find_all('table')[1]
```

```python
kolom = tabel.find_all('th')[0:6]
```

```python
kolom
```

```python
text_kolom = [title.text.strip() for title in kolom]
text_kolom
```

```python
import pandas as pd

df = pd.DataFrame(columns = text_kolom)
df = df.drop('Judul', axis = 1)
```

```python
df
```

```python
colom_data = tabel.find_all('tr')
```

```python
td = tabel.find_all('i')
judul = [data.text.strip() for data in td]
judul
```

```python
for row in colom_data[1:]:
    row_data = row.find_all('td') 
    individual_row_data = [data.text.strip() for data in row_data]
    
    lenght = len(df)
    df.loc[lenght] = individual_row_data
```

```python
df['judul'] = judul
```

```python
df.head()
```

```python
df.drop('Referensi', axis = 1, inplace=True)
```

```python
df.head()
```

```python
huruf = []
for data in df.Puncak:
    try:
        float(data)
    except ValueError:
        huruf.append(data)
set(huruf)
```

```python
df.Puncak = df.Puncak.str.replace('19DM2','19').str.replace('2PS','2').str.replace('4TS3','4')
```

```python
df = df[['Peringkat','Puncak','judul','Penghasilan kotor global','Tahun']]
```

```python
karakter = ['$',',','.']
df['Penghasilan kotor global'] = df['Penghasilan kotor global'].apply(lambda x: ''.join(char for char in str(x) if char not in karakter))
```

```python
huruf = []
for data in df['Penghasilan kotor global']:
    try:
        float(data)
    except ValueError:
        huruf.append(data)
set(huruf)
```

```python
df['Penghasilan kotor global'] = df['Penghasilan kotor global'].str.replace('F1290000000','1290000000').str.replace('F81238764765','1238764765')
```

```python
df['Penghasilan kotor global'] = df['Penghasilan kotor global'].astype('int64')
```
